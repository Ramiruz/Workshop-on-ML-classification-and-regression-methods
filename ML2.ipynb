# Graficas para el caso de estudio



```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data= "..\WorkshopOnML\Daily Demand Forecasting Orders_Daily_Demand_Forecasting_Orders.csv"

ML_data = pd.read_csv(data, index_col="Week of the month (first week second third fourth or fifth week)")

ML_data
```

```python
plt.figure(figsize=(16,6))

sns.lineplot(data=ML_data)
```

```python
plt.figure(figsize=(14,6))
plt.title("Monthly order according to type")

sns.lineplot(data=ML_data['Order type A'], label="Order type A")
sns.lineplot(data=ML_data['Order type B'], label="Order type B")
sns.lineplot(data=ML_data['Order type C'], label="Order type C")

plt.ylabel("N of orders")
plt.xlabel("Week of the month")
```

```python
plt.figure(figsize=(14,6))
plt.title("Monthly orders according to the urgent")

sns.lineplot(data=ML_data['Non-urgent order'], label="Non-urgent order")
sns.lineplot(data=ML_data['Urgent order'], label="Urgent order")

plt.ylabel("N of orders")
plt.xlabel("Week of the month")
```

```python
plt.figure(figsize=(14,6))
plt.title("Monthly banking orders")

sns.lineplot(data=ML_data['Banking orders (1)'], label="Banking orders (1)")
sns.lineplot(data=ML_data['Banking orders (2)'], label="Banking orders (2)")
sns.lineplot(data=ML_data['Banking orders (3)'], label="Banking orders (3)")

plt.ylabel("N of orders")
plt.xlabel("Week of the month")
```

```python
sns.kdeplot(data=ML_data['Target (Total orders)'], shade=True)
```

```python
sns.regplot(x=ML_data['Day of the week (Monday to Friday)'], y=ML_data['Target (Total orders)'])
```
